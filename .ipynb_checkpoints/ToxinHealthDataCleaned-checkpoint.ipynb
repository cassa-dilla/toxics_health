# Importing Toxin Release Inventory Data (2020-2022)

Source: https://www.epa.gov/toxics-release-inventory-tri-program/tri-basic-data-files-calendar-years-1987-present

Data Dictionary: https://www.epa.gov/system/files/documents/2022-10/TRI_Basic_Data_File_Documentation_RY2021_102522.pdf

```python
# Initialize an empty dictionary to store DataFrames for each year
toxins_data = {}

# Loop through years from 2020 to 2022
for year in range(2020, 2023):
    # Construct the file path for the CSV file for the current year
    file_path = f'ToxicsReleaseInventory(TRI)ProgramBasicDataFiles/{year}_us.csv'

    # Read the CSV file into a DataFrame
    df = pd.read_csv(file_path)

    # Set '1. YEAR' and '4. FACILITY NAME' as multiple indices
    df.set_index(['1. YEAR', '4. FACILITY NAME'], inplace=True)

    # Store the DataFrame in the dictionary with the year as the key
    toxins_data[year] = df

# Now, toxins_data is a dictionary where keys are years and values are corresponding DataFrames
# You can access DataFrames for individual years like toxins_data[2012], toxins_data[2013], etc.
```

```python
# Initialize an empty list to store DataFrames with aligned columns
aligned_dfs = []

# Extract reference columns from the first DataFrame
reference_columns = list(toxins_data[list(toxins_data.keys())[0]].columns)

# Iterate through each year and its corresponding DataFrame
for year, df in toxins_data.items():
    # Make a copy of the DataFrame to avoid modifying the original
    df_copy = df.copy()

    # Add any missing columns and fill them with NaN values
    missing_columns = set(reference_columns) - set(df.columns)
    for column in missing_columns:
        df_copy[column] = None  # Or fill with any appropriate value, like NaN

    # Reorder columns to match the reference columns order
    df_copy = df_copy[list(reference_columns)]

    # Append the DataFrame to the list
    aligned_dfs.append(df_copy)

# Concatenate all aligned DataFrames row-wise
concatenated_df = pd.concat(aligned_dfs)

# Display the concatenated DataFrame
concatenated_df.head()
```

```python
concatenated_df.columns
```

# Cleaning Toxin Data

```python
# Find unique values in the '47. UNIT OF MEASURE' column
unique_values = concatenated_df['47. UNIT OF MEASURE'].unique()

# Print the unique values
print(unique_values)
```

```python
# Group the data by year, county, and state, and calculate the sum of on-site release totals
grouped_df = concatenated_df.groupby(['1. YEAR', '8. ST', '7. COUNTY'])['62. ON-SITE RELEASE TOTAL'].sum().reset_index()

# Sort the DataFrame by year, state, and county
toxin_df = grouped_df.sort_values(by=['1. YEAR', '8. ST', '7. COUNTY'])

# Display the DataFrame showing the sum of on-site release totals for each county for each year, sorted by year, state, and county
toxin_df.head()
```

```python
# Rename all counties to include the word 'COUNTY' at the end for cleaning and standardization of county labels across datasets
toxin_df['7. COUNTY'] = toxin_df['7. COUNTY'].apply(lambda x: x.strip() + ' COUNTY')
```

```python
#Rename year column to standardize with year column in health dataframe
toxin_df.rename(columns={'1. YEAR': 'Year'}, inplace=True)

# Modify '7. COUNTY' column conditionally to match formatting of county names in health data
toxin_df['7. COUNTY'] = toxin_df['7. COUNTY'].str.replace(' parish', '', case=False) \
                                             .str.replace(' CENSUS ARE', '', case=False) \
                                             .str.replace(' municipality', '', case=False) \
                                             .str.replace(' municipio', '', case=False) \
                                             .str.replace(' city', '(CITY)', case=True) \
                                             .str.replace(' borough', '', case=False) \

# Define a dictionary mapping state abbreviations to their full names
state_names = {
    'AK': 'Alaska', 'AL': 'Alabama', 'AR': 'Arkansas', 'AZ': 'Arizona', 'CA': 'California',
    'CO': 'Colorado', 'CT': 'Connecticut', 'DE': 'Delaware', 'FL': 'Florida', 'GA': 'Georgia',
    'HI': 'Hawaii', 'IA': 'Iowa', 'ID': 'Idaho', 'IL': 'Illinois', 'IN': 'Indiana',
    'KS': 'Kansas', 'KY': 'Kentucky', 'LA': 'Louisiana', 'MA': 'Massachusetts', 'MD': 'Maryland',
    'ME': 'Maine', 'MI': 'Michigan', 'MN': 'Minnesota', 'MO': 'Missouri', 'MS': 'Mississippi',
    'MT': 'Montana', 'NC': 'North Carolina', 'ND': 'North Dakota', 'NE': 'Nebraska', 'NH': 'New Hampshire',
    'NJ': 'New Jersey', 'NM': 'New Mexico', 'NV': 'Nevada', 'NY': 'New York', 'OH': 'Ohio',
    'OK': 'Oklahoma', 'OR': 'Oregon', 'PA': 'Pennsylvania', 'RI': 'Rhode Island', 'SC': 'South Carolina',
    'SD': 'South Dakota', 'TN': 'Tennessee', 'TX': 'Texas', 'UT': 'Utah', 'VA': 'Virginia',
    'VT': 'Vermont', 'WA': 'Washington', 'WI': 'Wisconsin', 'WV': 'West Virginia', 'WY': 'Wyoming', 'PR': 'Puerto Rico'
}

# Function to combine county, state abbreviation, and full state name
def combine_geo_qname(row):
    county = row['7. COUNTY'].upper()  # Convert county name to uppercase
    state_abbr = row['8. ST']
    full_state_name = state_names.get(state_abbr, state_abbr).upper()  # Use the full state name if abbreviation not found, and convert to uppercase
    return f"{county}, {full_state_name}"

# Apply the function to create the Geo_QNAME column
toxin_df['Geo_QNAME'] = toxin_df.apply(combine_geo_qname, axis=1)

# Remove white spaces or special characters from all columns
toxin_df = toxin_df.applymap(lambda x: x.strip() if isinstance(x, str) else x)

# Convert 'Year' column to categorical
toxin_df['Year'] = toxin_df['Year'].astype('category')

# Convert 'Geo_QNAME' column to categorical
toxin_df['Geo_QNAME'] = toxin_df['Geo_QNAME'].astype('category')

toxin_df.set_index(['Year', 'Geo_QNAME'], inplace=True)

# Display the modified DataFrame
toxin_df.head()
```

# Importing County Health Data (2020-2022)

```python
# Initialize an empty dictionary to store DataFrames for each year
health_data = {}

# Loop through years from 2020 to 2022
for year in range(2020, 2023):
    # Construct the file path for the CSV file for the current year
    file_path = f'SocialExplorerHealthData/HealthData{year}.csv'

    # Read the CSV file into a DataFrame
    df = pd.read_csv(file_path)

    # Rename annual columns based on the specified pattern
    columns_to_rename = {col: col.split(f'HD{year}_')[-1] for col in df.columns if f'HD{year}_' in col}
    df.rename(columns=columns_to_rename, inplace=True)

    columns_to_rename = {col: col.split(f'SE_')[-1] for col in df.columns if f'SE_' in col}
    df.rename(columns=columns_to_rename, inplace=True)

    # Add a new column for the year
    df['Year'] = year

    # Set the index to 'Year' and 'Geo_QNAME'
    df.set_index(['Year', 'Geo_QNAME'], inplace=True)

    # Store the DataFrame in the dictionary with the year as the key
    health_data[year] = df

# Now, health_data is a dictionary where keys are years and values are corresponding DataFrames
# You can access DataFrames for individual years like health_data[2012], health_data[2013], etc.
```

```python
# Get the columns present in the DataFrame for the first year
common_columns = set(health_data[min(health_data.keys())].columns)

# Iterate over the remaining years to find the intersection of columns
for year in health_data.keys():
    common_columns = common_columns.intersection(health_data[year].columns)

# Convert the set of common columns to a list
common_columns = list(common_columns)

# Filter each DataFrame in health_data to include only the common columns
for year in health_data.keys():
    health_data[year] = health_data[year][common_columns]

# Now, health_data contains only the common columns present throughout all of the years
```

```python
# Concatenate the DataFrames in the health_data dictionary into a single DataFrame
health_df = pd.concat(health_data.values())

# Display the combined DataFrame
health_df.head()
```

# Cleaning Health Data

```python
# Convert all values in the second index column to uppercase while keeping the first index column unchanged
health_df.index = pd.MultiIndex.from_tuples(
    [(first_level, second_level.upper()) for first_level, second_level in health_df.index],
    names=health_df.index.names
)

# Reset the index temporarily to access 'Geo_QNAME'
health_df.reset_index(inplace=True)

# Modify 'Geo_QNAME' column conditionally
health_df['Geo_QNAME'] = health_df['Geo_QNAME'].str.replace(' parish', '', case=False) \
                                             .str.replace(' CENSUS ARE', '', case=False) \
                                             .str.replace(' municipality', '', case=False) \
                                             .str.replace(' municipio', '', case=False) \
                                             .str.replace(' city', '(CITY)', case=True) \
                                             .str.replace(' borough', '', case=False)

# Remove white spaces or special characters from all columns
health_df = health_df.applymap(lambda x: x.strip() if isinstance(x, str) else x)

# Convert 'Year' column to categorical
health_df['Year'] = health_df['Year'].astype('category')

# Convert 'Geo_QNAME' column to categorical
health_df['Geo_QNAME'] = health_df['Geo_QNAME'].astype('category')

# Set the indices back
health_df.set_index(['Year', 'Geo_QNAME'], inplace=True)  # replace 'other_index' with your other index column(s)

health_df = health_df.dropna(axis=1, how='all')

# Display the modified DataFrame
health_df.head()
```

```python
import matplotlib.pyplot as plt
from pysal.explore import esda
import mapclassify

# Make a copy of merged_df called quantile_df
quantile_df = merged_df.copy()

# Convert 'Geo_FIPS' column to string in the quantile_df DataFrame
quantile_df['Geo_FIPS'] = quantile_df['Geo_FIPS'].astype(str)

# Remove leading zeros from 'GEOID' column in the counties GeoDataFrame
counties['GEOID'] = counties['GEOID'].str.lstrip('0')

# Merge your DataFrame with the county geometries based on county names
merged_data = counties.merge(quantile_df, left_on='GEOID', right_on='Geo_FIPS')

# Filter out rows corresponding to Alaska and Hawaii
merged_data = merged_data[~merged_data['8. ST'].isin(['AK', 'HI'])]

# Extract the attribute data you want to classify
data_to_classify = merged_data['62. ON-SITE RELEASE TOTAL']

# Compute Jenks natural breaks
nb = mapclassify.NaturalBreaks(data_to_classify, k=5)  # Specify the number of classes (k)

# Get the class boundaries
breaks = nb.bins

# Assign class labels to each feature
classifications = nb.yb

# Plotting the chloropleth map with county outlines and enhanced colors
fig, ax = plt.subplots(figsize=(50, 10))
merged_data.plot(column=classifications, cmap='YlOrRd', edgecolor='black', linewidth=0.5, legend=True, ax=ax)
ax.set_title('Chloropleth Map of On-Site Release Totals with County Outlines (Jenks Natural Breaks) - Excluding Alaska and Hawaii')
plt.show()
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
from shapely.geometry import Polygon
from matplotlib.patches import Patch

# Load the shapefiles for tribal lands and counties
tl_shapefile_path = 'Choropleth/tl_2023_us_aiannh/tl_2023_us_aiannh.shp'
tribal_lands = gpd.read_file(tl_shapefile_path)

# Define the bounding box for the continental USA
continental_bbox = (-124.848974, 24.396308, -66.93457, 49.384358)  # (minx, miny, maxx, maxy)

# Create a polygon from the bounding box
continental_polygon = gpd.GeoSeries([Polygon([
    (continental_bbox[0], continental_bbox[1]),
    (continental_bbox[0], continental_bbox[3]),
    (continental_bbox[2], continental_bbox[3]),
    (continental_bbox[2], continental_bbox[1]),
])], crs='EPSG:4326')

# Create a figure and axes
fig, ax = plt.subplots(figsize=(50, 10))

# Plot the counties
counties[counties.intersects(continental_polygon.unary_union)].plot(ax=ax, color='white', edgecolor='black', linewidth=0.5)

# Plot the choropleth map with county outlines and enhanced colors
merged_data.plot(column=classifications, cmap='YlOrRd', edgecolor='black', linewidth=0.5, legend=True, ax=ax)
ax.set_title('Choropleth Map of County On-Site Toxics Release Totals (Jenks Natural Breaks) with Tribal Land Overlay- Contiguous U.S.')

# Plot the tribal lands within the continental USA on top of the choropleth map and counties
tribal_lands[tribal_lands.intersects(continental_polygon.unary_union)].plot(ax=ax, color='lightblue', edgecolor='black', linewidth=0.5, alpha=0.5, label='Tribal Lands')

# Position the legend for the tribal lands
tribal_legend = ax.legend(loc='upper right', title='Tribal Lands', handles=[Patch(color='lightblue', label='Tribal Lands')])

plt.show()
```

#### HistGradientBoostingClassifier

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.compose import ColumnTransformer
from sklearn.experimental import enable_hist_gradient_boosting
from sklearn.ensemble import HistGradientBoostingClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

filtered_df = merged_df

# Calculate the 50th percentile of '62. ON-SITE RELEASE TOTAL'
top_percent_threshold = filtered_df['62. ON-SITE RELEASE TOTAL'].quantile(0.75)

# Create a new binary column indicating if the value is in the top %
filtered_df['top_percent'] = (filtered_df['62. ON-SITE RELEASE TOTAL'] >= top_percent_threshold).astype(int)

# Split the filtered data into training (70%) and testing (30%)
df_train, df_test = train_test_split(filtered_df, test_size=0.3, random_state=42)

# Define target column
target = 'top_percent'

# Define features (excluding the target column)
features = df_train.columns[df_train.columns != target]

# Create X_train with specific columns and remove NaN values
X_train = df_train[features]
y_train = df_train[target]

X_test = df_test[features]
y_test= df_test[target]

# Preprocessing
categorical_columns = X_train.select_dtypes(include=['object']).columns
quantitative_columns = X_train.select_dtypes(include=['int', 'float']).columns

# Preprocessing with ColumnTransformer
col_transformer = ColumnTransformer([
    ('one_hot_encode', OneHotEncoder(sparse=False, handle_unknown='ignore'),
     categorical_columns),
    ('standard_scaler', StandardScaler(), quantitative_columns)
], remainder='passthrough')

# Initialize the HistGradientBoostingClassifier pipeline
pipeline_hist = make_pipeline(
    col_transformer,
    HistGradientBoostingClassifier()
)

pipeline_hist.steps
```

```python
# Fit the classifier to the training data
pipeline_hist.fit(X_train, y_train)

# Make predictions on the test data
y_pred = pipeline_hist.predict(X_test)
```

```python
# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
from sklearn.metrics import confusion_matrix, precision_score, recall_score

# Precision (P): Precision is the ratio of true positive predictions to the
# total number of positive predictions.
# Recall (R): also known as sensitivity or true positive rate, is the ratio
# of true positive predictions to the total number of actual positive instances.

# Assuming y_test contains the true labels and pred_top contains the predicted labels
conf_matrix = confusion_matrix(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)

# Print confusion matrix
confusion_matrix_df = pd.DataFrame(conf_matrix, columns=pipeline_hist.classes_, index=pipeline_hist.classes_)
print("Confusion Matrix:")
confusion_matrix_df
```
